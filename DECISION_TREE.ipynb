```python
df['alcohol'] = df.alcohol.replace(0,df.alcohol.mean())
df['total sulfur dioxide'] = df['total sulfur dioxide'].replace(0,df['total sulfur dioxide'].median())
df['free sulfur dioxide'] = df['free sulfur dioxide'].replace(0,df['free sulfur dioxide'].median())
df['residual sugar'] = df['residual sugar'].replace(0,df['residual sugar'].median())
df['citric acid'] = df['citric acid'].replace(0,df['citric acid'].median())
df['volatile acidity'] = df['volatile acidity'].replace(0,df['volatile acidity'].median())
df['chlorides'] = df.chlorides.replace(0,df.chlorides.median())
```

replaced 0 values with some data

```python
#separating dependent & independent feature
X = df.iloc[:,:-1]
y=df.iloc[:,-1]
```

OUTLIER REMOVAL
# **APPROACH 1: STANDARDISATION**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled)
# X_scaled.columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']
```

```python
X_scaled.columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']
```

```python
 fig,ax = plt.subplots(figsize=(18,15))
 sns.boxplot(data = X_scaled,ax=ax)
 plt.savefig('standardised-boxplot.jpg')
```

# **SECOND APPROACH : QUANTILE APPROACH**

```python
X_scaled.reset_index(drop=True,inplace=True)
Y_scaled = y
```

```python
Y_scaled = pd.DataFrame(Y_scaled)
Y_scaled.reset_index(drop=True,inplace=True)
```

```python
q = X_scaled['chlorides'].quantile(0.95)
mask = X_scaled['chlorides'] < q
datanew = X_scaled[mask]
Y_scaled = Y_scaled[mask]
```

```python
 fig,ax = plt.subplots(figsize=(18,15))
 sns.boxplot(data = datanew,ax=ax)
 plt.savefig('standardised-boxplot.jpg')
```

```python
 q = datanew['residual sugar'].quantile(0.95)
 mask = datanew['residual sugar']<q
 datanew2 = datanew[mask]
 Y_scaled = Y_scaled[mask]
```

```python
 q = datanew2['sulphates'].quantile(0.95)
 mask = datanew2['sulphates']<q
 datanew3 = datanew2[mask]
 Y_scaled = Y_scaled[mask]
```

```python
 fig,ax = plt.subplots(figsize=(18,15))
 sns.boxplot(data = datanew3,ax=ax)
 plt.savefig('standardised-boxplot.jpg')
```

## **datanew3 have much better with very few and robust outliers for model training**

# **MODEL TRAINING**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(datanew3,Y_scaled)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

# MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
